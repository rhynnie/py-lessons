```python
numberone=('','один','два','три','четыре','пять','шесть','семь','восемь','девять','десять','одинадцать', 'двенадцать','тринадцать','четырнадцать','пятнадцать','шестнадцать','семнадцать', 'восемнадцать','девятнадцать') #выведение числа до 20 + единиц в двузначном числе (первое число пропущенно из-за того, что отсчёт начинается с 0)
numbertwo=('','десять','двадцать','тридцать','сорок','пятьдесят','шестьдесят','семьдесят','восемьдесят','девяносто') #выведение десятков (первое число пропущенно из-за того, что отсчёт начинается с 0)
guess=int(input("Введите число от 0 до 99: ")) #ввод числа
if guess==0:print('ноль') #если введённое число равно нулю, пишет ноль
elif type(guess)=="str":print("Вы должны ввести число") #если введённое число - строка, пишет, что нужно ввести число (но так как мы сразу обозначили тип данных integer, он будет выдавать ошибку)
elif guess<0 or guess>99:print("Введённое число не входит в заданные рамки" )#если введённое число выходит за рамки возможных, программа сообщает об этом
elif guess>0 and guess<20:print(numberone[guess]) #при введении числа меньше 20, выводит число из первого списка по номеру вводимого числа
else: print(numbertwo[guess//10],numberone[guess%10]) #при введении числа больше 20, выводит первую часть числа из первого списка и вторую чать числа из второго
```
