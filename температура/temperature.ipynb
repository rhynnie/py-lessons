```python
scale = int(input("Какую процедуру необходимо выполнить?\n1) Перевод цельсия в фаренгейт\n2) Перевод фаренгейта в цельсий\n")) #выясняем, какую шкалу переводить и зашифровываем шкалы под номерами
```

```python
if scale == 1: #если перевод из цельсия
  c = float(input("Введите градусы цельсия: ")) #вводим градусы цельсия
  if c < -273.15: #если введённое число меньше абсолютного нуля температуры
    print("Данной температуры не существует") #такого не бывает
  else:
    ftemp = c*1.8+32 #переводим цельсий в фаренгейт
    print(f"Температура {c} градусов цельсия равна {ftemp} градусам фаренгейта") #выводим перевод градусов
elif scale == 2: #если перевод из фаренгейта
  f = float(input("Введите градусы фаренгейта: ")) #вводим градусы фаренгейта
  if f < -459.67: #если введённое число меньше абсолютного нуля температуры
    print("Данной температуры не существует") #такого не бывает
  else:
    ctemp = (f-32)*1.8 #переводим фаренгейт в цельсий
    print(f"Температура {f} градусов фаренгейта равна {ctemp} градусам цельсия") #выводим перевод градусов
else: #если вводим неправильное число выполняемой процедуры
  print ("Данное число не входит в заданные параметры") #завершаем
```
